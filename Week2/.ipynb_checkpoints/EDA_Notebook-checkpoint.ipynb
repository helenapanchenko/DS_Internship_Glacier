# Profit Analysis

```python
pink_profit = all_data[all_data['Company'] == 'Pink Cab'].groupby(all_data['Date of Travel'].dt.year)['Profit'].sum()
yellow_profit = all_data[all_data['Company'] == 'Yellow Cab'].groupby(all_data['Date of Travel'].dt.year)['Profit'].sum()
all_data['Profit per KM'] = all_data['Profit'] / all_data['KM Travelled']
pink_profit_per_KM = all_data[all_data['Company'] == 'Pink Cab'].groupby(all_data['Date of Travel'].dt.year)['Profit per KM'].mean()
yellow_profit_per_KM = all_data[all_data['Company'] == 'Yellow Cab'].groupby(all_data['Date of Travel'].dt.year)['Profit per KM'].mean()

fig, axes = plt.subplots(2, 1, figsize=(8, 10))
axes = axes.flatten()

bar_width = 0.35

x_pink = np.arange(len(pink_profit))
x_yellow = np.arange(len(yellow_profit)) + bar_width

axes[0].bar(x_pink, pink_profit, width=bar_width, color='pink', alpha=0.7, label='Pink Cab')
axes[0].bar(x_yellow, yellow_profit, width=bar_width, color='yellow', alpha=0.7, label='Yellow Cab')
axes[0].set_title('Annual Profit of Cabs')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Profit')
axes[0].set_xticks(x_pink + bar_width / 2)
axes[0].set_xticklabels(pink_profit.index)
axes[0].legend()
axes[0].get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{int(x):,}'))

axes[1].bar(x_pink, pink_profit_per_KM, width=bar_width, color='pink', alpha=0.7, label='Pink Cab')
axes[1].bar(x_yellow, yellow_profit_per_KM, width=bar_width, color='yellow', alpha=0.7, label='Yellow Cab')
axes[1].set_title('Average Annual Profit per KM of Cabs')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Profit per KM')
axes[1].set_xticks(x_pink + bar_width / 2)
axes[1].set_xticklabels(pink_profit_per_KM.index)
axes[1].legend()

plt.tight_layout()
```

Conclusion: 
1. Yellow Cab's annual profit is about 8 times Pink Cab's annual profit.
2. Yellow Cab's average profit per KM is about 2.5 times Pink Cab's profit per KM.

```python
city_company_counts = all_data.groupby(['City', 'Company']).size().unstack(fill_value=0)

# Calculate the total number of rides for each city
city_total_rides = city_company_counts.sum(axis=1)

# Sort the cities based on the total number of rides in descending order
sorted_cities = city_total_rides.sort_values(ascending=False).index

# Reorder the rows of the DataFrame based on the sorted cities
city_company_counts_sorted = city_company_counts.loc[sorted_cities]

# Plot the stacked bar plot with specified colors
city_company_counts_sorted.plot(kind='bar', stacked=True, figsize=(10, 6), color=['pink', 'yellow'])

# Add labels and title
plt.xlabel('City')
plt.ylabel('Number of Rides')
plt.title('Number of Rides by City and Company')

# Show the plot
plt.legend(title='Company')
```

Conclusion: The ratio of total number of rides varies across cities, but Yellow Cab dominates in most.

```python
pink_data = all_data[all_data['Company'] == 'Pink Cab'].copy()
yellow_data = all_data[all_data['Company'] == 'Yellow Cab'].copy()

pink_data['Year'] = pink_data['Date of Travel'].dt.year
pink_data['Month'] = pink_data['Date of Travel'].dt.month

yellow_data['Year'] = yellow_data['Date of Travel'].dt.year
yellow_data['Month'] = yellow_data['Date of Travel'].dt.month

pink_monthly_counts = pink_data.groupby(['Year', 'Month']).size().unstack(fill_value=0)
yellow_monthly_counts = yellow_data.groupby(['Year', 'Month']).size().unstack(fill_value=0)

pink_monthly_profit = pink_data.groupby(['Year', 'Month'])['Profit'].sum()
yellow_monthly_profit = yellow_data.groupby(['Year', 'Month'])['Profit'].sum()

fig_pink, axes_pink = plt.subplots(1, 3, figsize=(18, 6), sharey=True, constrained_layout=True)
fig_yellow, axes_yellow = plt.subplots(1, 3, figsize=(18, 6), sharey=True, constrained_layout=True)

fig_pink.text(0.5, 1.05, 'Transaction and Profit Distribution throughout the Year', fontsize=16, ha='center')
fig_pink.suptitle('Pink Cab', fontsize=16)
fig_yellow.suptitle('Yellow Cab', fontsize=16)

for i, year in enumerate(pink_monthly_counts.index):
    ax = axes_pink[i]
    pink_monthly_counts.loc[year].plot(kind='bar', ax=ax, color='pink', alpha=0.8)
    ax.set_xlabel(f'{year}')
    ax.set_ylabel('Number of Rides')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
    ax_twin = ax.twinx()
    ax_twin.plot(range(0, 12),pink_monthly_profit.loc[year], marker='o', color='red', label='Profit')
    ax_twin.legend(loc='upper right')
    if i == len(axes_pink) - 1:
        ax_twin.tick_params(axis='y', labelright=True)
        ax_twin.set_ylabel('Profit')
    else:
        ax_twin.tick_params(axis='y', labelright=False)

for i, year in enumerate(yellow_monthly_counts.index):
    ax = axes_yellow[i]
    yellow_monthly_counts.loc[year].plot(kind='bar', ax=ax, color='yellow', alpha=0.8)
    ax.set_xlabel(f'{year}')
    ax.set_ylabel('Number of Rides')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
    ax_twin = ax.twinx()
    ax_twin.plot(range(0, 12), yellow_monthly_profit.loc[year], marker='o', color='orange', label='Profit')
    ax_twin.legend(loc='upper right')
    if i == len(axes_yellow) - 1: 
        ax_twin.tick_params(axis='y', labelright=True)
        ax_twin.set_ylabel('Profit')
    else:
        ax_twin.tick_params(axis='y', labelright=False)
```

Conclusion: 
1. The number of rides for both companies increases throughout the year, with a peak in December and a rise in January, likely due to holidays.
2. Pink Cab likely maintains a consistent profit margin year-round, while Yellow Cab sets a higher margin in the May-June period and a lower margin in the July-August period.

## Hypothesis
1. There is any seasonality in number of customers using the cab service.
2. Amount of money person spends on taxi annualy depends on income.
3. Gender has effect on company profit.
4. Age has effect on company profit.
5. Holidays impact the number of rides.

### Hypothesis 1.
H0: The number of transactions doesn't depend on the month of the year.

H1: The number of transactions depends on the month of the year.

```python
# Data
hyp1_data = cabs.loc[:,['Date of Travel', 'Transaction ID']]
hyp1_data['Month'] = hyp1_data['Date of Travel'].dt.month
monthly_transactions = hyp1_data.groupby('Month')['Transaction ID'].count().reset_index()
monthly_transactions.columns = ['Month', 'Transaction Count']


# The Chi-Square Test
expected_frequencies = [sum(monthly_transactions['Transaction Count']) / 12] * 12
chi2_stat, p_val = stats.chisquare(monthly_transactions['Transaction Count'], f_exp=expected_frequencies)

print(f'Chi2 Stat: {chi2_stat}')
print(f'p-value: {p_val}')

# Make a Decision
alpha = 0.05   # Significance Level
if p_val < alpha:
    print("Reject the null hypothesis (H0): The number of transactions depends on the month of the year.")
else:
    print("Fail to reject the null hypothesis (H0): The number of transactions does not depend on the month of the year.")
```

### Hypothesis 2.
H0: Amount of money person spends on taxi annualy doesn't depend on income.

H1: Amount of money person spends on taxi annualy depends on income.

```python
# Data
hyp1_data = cabs.loc[:,['Date of Travel', 'Transaction ID']]
hyp1_data['Month'] = hyp1_data['Date of Travel'].dt.month
monthly_transactions = hyp1_data.groupby('Month')['Transaction ID'].count().reset_index()
monthly_transactions.columns = ['Month', 'Transaction Count']


# The Chi-Square Test
expected_frequencies = [sum(monthly_transactions['Transaction Count']) / 12] * 12
chi2_stat, p_val = stats.chisquare(monthly_transactions['Transaction Count'], f_exp=expected_frequencies)

print(f'Chi2 Stat: {chi2_stat}')
print(f'p-value: {p_val}')

# Make a Decision
alpha = 0.05   # Significance Level
if p_val < alpha:
    print("Reject the null hypothesis (H0): The number of transactions depends on the month of the year.")
else:
    print("Fail to reject the null hypothesis (H0): The number of transactions does not depend on the month of the year.")
```

### Hypothesis 3.
H0: Gender has no effect on company profit.

H1: Gender has an effect on company profit.

```python
# Perform the two-sample t-test
male_profits = all_data[all_data['Gender'] == 'Male']['Profit']
female_profits = all_data[all_data['Gender'] == 'Female']['Profit']

t_stat, p_value = stats.ttest_ind(male_profits, female_profits, equal_var=False)

print(f't-statistic: {t_stat}')
print(f'p-value: {p_value}')

# Determine the result of the hypothesis test
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0): Gender has an effect on company profit.")
else:
    print("Fail to reject the null hypothesis (H0): Gender has no effect on company profit.")
```

### Hypothesis 4.
H0: Age doesn't affect company profit.

H1: Age affects company profit.

```python
bins = [0, 18, 24, 34, 44, 54, 64, 100]
labels = ['Under 18', '18-24', '25-34', '35-44', '45-54', '55-64', '65 and above']
all_data['AgeGroup'] = pd.cut(all_data['Age'], bins=bins, labels=labels)

# Perform ANOVA test
grouped_data = [group['Profit'].values for name, group in all_data.groupby('AgeGroup', observed=False)]
f_stat, p_value = stats.f_oneway(*grouped_data)

print(f'F-statistic: {f_stat}')
print(f'p-value: {p_value}')

# Determine the result of the hypothesis test
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0): Age affects company profit.")
else:
    print("Fail to reject the null hypothesis (H0): Age doesn't affect company profit.")
```

### Hypothesis 5
H0: Holidays do not effect amount of transactions.

H1: Holidays effect amount of transactions.

```python
holidays = pd.read_csv('US_Holidays.csv')
holidays.head()
```

```python
holidays['Date'].dtype
```

```python
holidays['Date'] = pd.to_datetime(holidays['Date'], errors='coerce').dt.date
```

```python
holidays['Date'].min()
```

```python
holidays['Date'].max()
```

```python
hyp5_data = all_data[['Transaction ID','Date of Travel']]
hyp5_data['Date of Travel'] = pd.to_datetime(hyp5_data['Date of Travel'], errors='coerce').dt.date
daily_transactions = hyp5_data.groupby('Date of Travel').size().reset_index(name='Transaction Count')
daily_transactions['Holiday'] = daily_transactions['Date of Travel'].isin(holidays['Date'])

holiday_transactions = daily_transactions[daily_transactions['Holiday'] == True]['Transaction Count']
non_holiday_transactions = daily_transactions[daily_transactions['Holiday'] == False]['Transaction Count']

# Perform an independent samples t-test
t_stat, p_value = stats.ttest_ind(holiday_transactions, non_holiday_transactions, equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Holidays do affect the number of transactions.")
else:
    print("Fail to reject H0: Holidays do not affect the number of transactions.")
```
